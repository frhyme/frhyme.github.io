```python
import numpy as np 
import matplotlib.pyplot as plt
"""
-            . 
-  step   -1, 0, 1      ,        . 

"""
plt.figure(figsize=(15, 4))
for i in range(0, 5):
    xs = np.array([np.random.choice([-1,0, 1]) for i in range(0, 100)])
    cum_xs = np.cumsum(xs)
    plt.plot(np.arange(0, len(xs)), cum_xs, linestyle='--', label='random_walk_{}'.format(i))
plt.legend()
plt.show()
```

```python
# import mnist data
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import fetch_mldata
mnist = fetch_mldata("MNIST original")
X = mnist.data / 255.0 # np.array (70000, 784)
y = mnist.target # np.array (70000,)

# data  ,      . 
def sample_n(X, y, n):
    selected_rows = [int(70000/n*i) for i in range(0, n)]
    new_X = X[selected_rows]
    new_y = y[selected_rows]
    return new_X, new_y

from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.manifold import Isomap
def dim_reduction_and_draw(TSNE_or_PCA, X, y, output_file):
    new_X = TSNE_or_PCA(n_components=2).fit_transform(X)
    f = plt.figure()
    plt.scatter(new_X[:, 0], new_X[:, 1], c=y, alpha=0.7)
    plt.title(output_file)
    plt.axis('off')
    plt.legend()
    plt.show()
    plt.savefig("../../assets/images/markdown_img/"+output_file)

new_X, new_y = sample_n(X, y, 200)
dim_reduction_and_draw(TSNE, new_X, new_y, "TSNE1.svg")
dim_reduction_and_draw(TSNE, new_X, new_y, "TSNE2.svg")
dim_reduction_and_draw(PCA, new_X, new_y, "PCA1.svg")
dim_reduction_and_draw(PCA, new_X, new_y, "PCA2.svg")
dim_reduction_and_draw(Isomap, new_X, new_y, "Isomap1.svg")
dim_reduction_and_draw(Isomap, new_X, new_y, "Isomap2.svg")
```

```python
import time
start = time.time()
dim_reduction_and_draw(PCA, new_X, new_y, "PCA1.svg")
print(time.time() - start)
start = time.time()
dim_reduction_and_draw(TSNE, new_X, new_y, "PCA1.svg")
print(time.time() - start)
```
